# Data Exploration - IMDB Dataset
Reproduces Figure 2 from the paper: token length distribution of title vs plot.

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from utils.tokenizer import count_tokens

df = pd.read_csv('movies_full.csv')
print(f'Dataset: {len(df)} movies')
print(f'Languages: {df["language"].nunique()} unique')
print(f'Genres: {df["genre"].nunique()} unique')
df.head()
```

```python
df['title_tokens'] = df['title'].apply(count_tokens)
df['plot_tokens'] = df['plot'].apply(count_tokens)

print('Title tokens: mean={:.1f}, median={:.1f}, min={}, max={}'.format(
    df['title_tokens'].mean(), df['title_tokens'].median(),
    df['title_tokens'].min(), df['title_tokens'].max()))
print('Plot tokens:  mean={:.1f}, median={:.1f}, min={}, max={}'.format(
    df['plot_tokens'].mean(), df['plot_tokens'].median(),
    df['plot_tokens'].min(), df['plot_tokens'].max()))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df['title_tokens'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Input Length (Title)')
axes[0].set_ylabel('Occurrences')
axes[0].set_title('(a) Title Token Distribution')

axes[1].hist(df['plot_tokens'], bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Input Length (Plot)')
axes[1].set_ylabel('Occurrences')
axes[1].set_title('(b) Plot Token Distribution')

plt.tight_layout()
plt.savefig('../results/plots/fig2_token_distribution.png', dpi=150)
plt.show()
```

```python
# Language and genre distributions
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df['language'].value_counts().head(10).plot.bar(ax=axes[0], color='steelblue', edgecolor='black')
axes[0].set_title('Top 10 Languages')
axes[0].set_ylabel('Count')

df['genre'].value_counts().head(10).plot.bar(ax=axes[1], color='coral', edgecolor='black')
axes[1].set_title('Top 10 Genres')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
